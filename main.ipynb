# Run data preparation using the SONAR embeddings.

```python
import os
from subprocess import run, CalledProcessError

# Define the output directory for data
output_dir = "prepared_data"

# Ensure the 'large_concept_model' directory exists
lcm_script_path = os.path.join("large_concept_model", "scripts", "prepare_wikipedia.py")
if not os.path.exists(lcm_script_path):
    raise FileNotFoundError(f"Script not found: {lcm_script_path}")

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Run data preparation script
try:
    run(["python", lcm_script_path, output_dir], check=True)
    print(f"Data prepared in: {output_dir}")
except CalledProcessError as e:
    print(f"Error running data preparation script: {e}")
    print("Full error message:")
    print(e.output)
```
